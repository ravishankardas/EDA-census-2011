```python
int_df = pd.merge(rslt_df, fin_data2, how ='inner', on =['State'])
```

```python
int_df['only_one'] = int_df['TOT_P'] - int_df['S_persons']
```

```python
int_df['exactly_two'] = int_df['S_persons'] - int_df['T_persons']
```

```python
int_df.pop('S_persons')
```

```python
v1 = sum(int_df['TOT_P'].to_list())
v2 = sum(int_df['TOT_P'].to_list())
v3 = sum(int_df['only_one'].to_list())
v4 = sum(int_df['exactly_two'].to_list())
```

```python
int_df.loc[len(int_df.index)] = ['total',v1 , v2,v3,v4] 
```
